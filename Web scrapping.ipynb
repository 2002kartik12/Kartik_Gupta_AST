```python
result=soup.find('ul',class_='css-zu9cdh eu4oa1w0')
result1=result.find_all('li',class_='css-5lfssm eu4oa1w0')
print(result1)
```

```python
import numpy as np
import re
```

```python
pages=np.arange(1,2)
all_salaries = []
for page in pages:
    for i in result1:
        df=pd.DataFrame(columns = ['Job Title','Company Name','Location','Salary'])
        #job title
        try:
            job_title=i.find('h2',class_='jobTitle css-mr1oe7 eu4oa1w0')
            job_title=job_title.text
            print('job_title = ',job_title)
        except Exception as e:
            job_title = 'NOT AVAILABLE'
            pass

        #company name
        try:
            company_name=i.find('span',class_='css-1x7z1ps eu4oa1w0')
            company_name=company_name.text
            print('Company Name = ',company_name)
        except Exception as e:
            company_name='NOT AVAILABLE'

        #location
        try:
            location=i.find('div',class_='css-t4u72d eu4oa1w0')
            location=location.text
            print('location = ',location)
        except Exception as e:
            location='NOT AVAILABLE'
        try:
            sal = i.find('div', class_='css-1ihavw2 eu4oa1w0').text
        except Exception as e:
            print('Salary not available')
       
        print('\n')

        df1=pd.DataFrame([[job_title,company_name,location,sal]],columns=['Job Title','Company Name','Location','Salary'])
        df=pd.concat([df,df1],ignore_index=True)
        print(df)
        print("\n")

        job_data = {
            'Job Title': job_title,
            'Company Name': company_name,
            'Location': location,
            'Salary': sal
        }

        # Insert the data into MongoDB
        collection.insert_one(job_data)

        print("Data inserted into MongoDB:")
        print(job_data)
        print("\n")
    driver.execute_script("window.scrollTo(0,(document.body.scrollHeight))")
    driver.find_element(By.XPATH,'//*[@id="jobsearch-JapanPage"]/div/div[5]/div[1]/nav/ul/li[3]/a').click()
```

```python
if all_salaries:
    # Calculate the average salary using NumPy
    average_salary = np.mean(all_salaries)

    # Print the average salary
    print('Average Salary:', average_salary)
else:
    print('No valid salaries found.')
```
