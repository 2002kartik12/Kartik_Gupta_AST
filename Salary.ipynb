```python
result=soup.find('ul',class_='css-zu9cdh eu4oa1w0')
result1=result.find_all('li',class_='css-5lfssm eu4oa1w0')
print(result1)
```

```python
import numpy as np
import re
```

```python
pages=np.arange(1,2)
all_salaries = []
for page in pages:
    for i in result1:
        try:
            sal = i.find('div', class_='css-1ihavw2 eu4oa1w0').text
            sal = (sal.replace(',', ''))  
            sal = (sal.replace('', ''))
            sal = (sal.replace('month', ''))
            sal = (sal.replace('year', ''))
            sal = (sal.replace('a', ''))
            salary_numeric = [float(match.group()) for match in re.finditer(r'\b\d+(\.\d+)?\b', sal)]
            #salary_numeric = [int(i) for i in sal.replace('-', '').split() if i.isdigit()]
            #salary_numeric = [int(num.replace(',', '')) for num in sal.replace(' ', '').split() if num.isdigit()]
            print(salary_numeric)
            if salary_numeric:
                # Append the extracted salaries to the list
                all_salaries.extend(salary_numeric)

                # Print the extracted salaries
                print('Salaries = ', salary_numeric)
            else:
                print('No valid salaries found')
                    
        except Exception as e:
            print('Salary not available')
       
        print('\n')

    
    driver.execute_script("window.scrollTo(0,(document.body.scrollHeight))")
    driver.find_element(By.XPATH,'//*[@id="jobsearch-JapanPage"]/div/div[5]/div[1]/nav/ul/li[3]/a').click()
```

```python
if all_salaries:
    # Calculate the average salary using NumPy
    average_salary = np.mean(all_salaries)

    # Print the average salary
    print('Average Salary:', average_salary)
else:
    print('No valid salaries found.')
```
